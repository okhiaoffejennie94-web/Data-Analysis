```python
import pandas as pd
```

```python
crime = pd.read_csv('crime.csv',encoding='latin1')
crime
```

```python
crime.isnull()
```

```python
crime.isna()
```

```python
crime.isnull().sum()
```

```python
crime=crime.fillna('')
crime
```

```python
crime = crime.drop(['Location', 'YEAR', 'MONTH', 'HOUR'], axis = 1)
crime
```

```python
crime
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(crime['DISTRICT'], crime['REPORTING_AREA'], color = 'red', marker = 'd')
```

```python
crime = crime.dropna()
crime_district = crime [['DISTRICT', 'REPORTING_AREA']]
crime_district.sort_values(by = 'REPORTING_AREA', ascending=False)
crime_district_top20 = crime_district.head(20)
crime_district_top20
```

```python
plt.plot(crime_district_top20['DISTRICT'], crime_district_top20['REPORTING_AREA'], color = 'red', marker = 'd')

plt.xlabel('DISTRICT')
plt.ylabel('REPORTING_AREA')
plt.title('Crime District Top 20')
plt.show()
```
